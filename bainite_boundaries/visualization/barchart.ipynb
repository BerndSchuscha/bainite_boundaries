```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Apply a Seaborn style
sns.set_theme(style="whitegrid")

# Data
criteria = ['Low', 'Mid', 'High']
isothermal = [0, 2, 20]
continuous_cooled = [0, 60, 79]

# Bar width and positions
bar_width = 0.4
x = np.arange(len(criteria))

# Create the plot
fig, ax = plt.subplots(figsize=(10, 5))
bars1 = ax.bar(
    x - bar_width / 2, isothermal, bar_width, label='Isothermal',
    color=sns.color_palette("pastel")[0], edgecolor='black', linewidth=1, alpha=1
)
bars2 = ax.bar(
    x + bar_width / 2, continuous_cooled, bar_width, label='Continuous Cooled',
    color=sns.color_palette("pastel")[1], edgecolor='black', linewidth=1, alpha=1
)

# Add values on top of the bars
for bar in bars1:
    height = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width() / 2, height + 2, f'{int(height)}', 
        ha='center', va='bottom', fontsize=15, fontweight='bold', color='black'
    )

for bar in bars2:
    height = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width() / 2, height + 2, f'{int(height)}', 
        ha='center', va='bottom', fontsize=15, fontweight='bold', color='black'
    )

# Labels and title
ax.set_xlabel('Acceptence part of the search space / %', fontsize=15, labelpad=10)
#ax.set_ylabel('acceptence rate / %', fontsize=15, labelpad=10)
ax.set_xticks(x)
ax.set_xticklabels(criteria, fontsize=15)
ax.legend(fontsize=12, title='Condition', title_fontsize=12, loc='upper left')

# Style tweaks
#ax.spines['top'].set_visible(False)
#ax.spines['right'].set_visible(False)
#ax.grid(axis='y', linestyle='--', alpha=0.7)
ax.set_axisbelow(True)
# Turn off all axis borders
for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.grid(False)
plt.yticks([])
ax = plt.gca()

# Turn on the x-axis spines (borders)
ax.spines['bottom'].set_visible(True)
# Improve layout
plt.tight_layout()
plt.show()
fig.savefig(f'accept_results.png', bbox_inches='tight', dpi=300)
```
