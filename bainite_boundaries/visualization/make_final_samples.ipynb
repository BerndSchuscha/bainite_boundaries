```python
import numpy as np
```

```python
#['C', 'SI', 'MN', 'CR', 'AL', 'V', 'Mo','T','Tcrit']

vmin=np.array([0.2,1,0,0,0,0,0,250,-1])
vmax=np.array([1,4,4,4,4,0.5,4,600,float(np.log10(25))])
num=int(1e6*3.5)
```

```python
sam=np.random.rand(num,9)
for k in range(9):
    sam[:,k]=sam[:,k]*(vmax[k]-vmin[k])+vmin[k]
```

```python
L=np.sum(sam[:,0:7],axis=1)<=10
```

```python
sam=sam[L,:]
sam=sam[:int(1e6),:]
np.savetxt('final_samples.txt',sam)
```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots(3,3)
ax=ax.flatten()
for k in range(9):
    if k ==8:
        ax[k].hist(10**(sam[:,k]))
    else:

        ax[k].hist((sam[:,k]))
```
