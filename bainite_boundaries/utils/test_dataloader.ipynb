```python
import bainite_boundaries.utils.load_data as ld
import matplotlib.pyplot as plt
import numpy as np

from tabulate import tabulate
```

```python
x, y,features=ld.load_dataset('final')
```

## Bainite

```python
x, y,features=ld.load_dataset('Bainite')
xx=np.zeros((len(x),np.shape(x[0])[1]))
for k in range(len(x)):
    xx[k,:]=x[k][0,:]
n=len(features)
fig,ax=plt.subplots(2,6,figsize=(15,5))
ax=ax.flatten()

for k in range(n):
    ax[k].hist(xx[:,k])
    ax[k].set_title(features[k])
fig.tight_layout()
```

```python
mmin=np.min(xx,axis=0).tolist()
mmax=np.max(xx,axis=0).tolist()
num=np.sum(xx>0,axis=0).tolist()

TT=[]
yy=[]
for k in range(len(x)):
    TT=TT+x[k][:,-1].tolist()
    yy=yy+y[k].tolist()
np.array(TT)

mmin=mmin+[float(np.min(TT))]
mmax=mmax+[float(np.max(TT))]
num=num+[int(len(TT))]
features.append('T') 
features[-2]='wper' 

mmin[-2]=np.min(np.sum(xx[:,:-1],axis=1)).tolist()
mmax[-2]=np.max(np.sum(xx[:,:-1],axis=1)).tolist()
num[-2]=np.size(np.sum(xx[:,:-1],axis=1))

print(tabulate([mmin,mmax,num],headers=features))

print(np.min(yy))
print(np.max(yy))
print(np.size(yy))
```
